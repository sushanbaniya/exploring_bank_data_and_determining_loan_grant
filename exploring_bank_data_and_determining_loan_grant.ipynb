```python
X = bank_data[['Income','CCAvg','Education','CD.Account']]
```

```python
print(X)
```

```python
Y = bank_data[['Personal.Loan']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
train_data_prediction = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
train_data_accuracy = accuracy_score(Y_train, train_data_prediction)
print('Training data accuracy is: ',train_data_accuracy)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(Y_test, test_data_prediction)
print('Testing data accuracy is: ', test_data_accuracy)
```
